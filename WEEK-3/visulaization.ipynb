```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Happiness.csv')
```

```python
dataset
```

```python
# Line graph
plt.figure(figsize=(8,5))  # Set the figure size to 8 inches wide and 5 inches tall
# Plot the Happiness Score against Overall Rank
plt.plot(dataset["Overall rank"],
         dataset["Score"],
         'g.-',                  # 'g' for green line, '.' for point markers, '-' for line
         label="Happiness Score")  # Label for the legend
# Plot GDP per capita against Overall Rank
plt.plot(dataset["Overall rank"],
         dataset["GDP per capita"],
         label="GDP per Capita")  # Label for the legend
# Plot Freedom to make life choices against Overall Rank
plt.plot(dataset["Overall rank"],
         dataset["Freedom to make life choices"],
         label="Freedom to Make Life Choices")  # Label for the legend
# Label the x-axis
plt.xlabel("Overall Rank")  # Lower rank means higher happiness
# Label the y-axis
plt.ylabel("Values")  # Y-axis shows the numerical values of the factors
# Add a title to the plot
plt.title("Comparison of Happiness Factors")  # Title of the graph
# Display the legend to identify the lines
plt.legend()
# Display the plot
plt.show()
```

```python
# LINE GRAPH
plt.figure(figsize=(12,5))  # Set the figure size to 12 inches wide and 5 inches tall
# Select the columns from index 2 onward to be plotted (excluding 'Country' and 'Overall rank')
columns_to_plot = dataset.columns[2:]
# Loop through each selected column and plot it against the Overall Rank
for column in columns_to_plot:
    plt.plot(dataset["Overall rank"],           # X-axis: Overall happiness rank of countries
             dataset[column],                   # Y-axis: Values of the selected factor (e.g., GDP, Freedom, etc.)
             marker='.',                        # Add a dot marker for each data point
             label=column)                      # Label each line with the column name for the legend
# Label the x-axis
plt.xlabel("Rank")  # X-axis represents the overall happiness rank
# Label the y-axis
plt.ylabel("Values")  # Y-axis represents the value of each contributing factor
# Add a title with bold weight and larger font size
plt.title("Comparison of Different Factors Over Years", fontdict={'fontweight': 'bold', 'fontsize': 22})
# Display the legend to differentiate the plotted lines
plt.legend()
# Show the final plot
plt.show()
```

```python
# Plot a histogram of the 'Perceptions of corruption' column
plt.hist(dataset['Perceptions of corruption'], color='#abcdef')  # Set custom color for the bars
# Label the x-axis
plt.xlabel('Perceptions of Corruption')  # Represents how people perceive corruption levels in their country
# Label the y-axis
plt.ylabel('Frequency')  # Number of countries falling into each corruption perception range
# Set the title of the histogram
plt.title('Distribution of Perceptions of Corruption')
# Display the histogram
plt.show()
```

```python
#PIE CHART
# Select the data row for Finland from the dataset
finland = dataset[dataset['Country or region'] == 'Finland'].iloc[0]
# Create a dictionary of key factors contributing to Finland's happiness score
factors = {
    'GDP': finland['GDP per capita'],                          # Economic output per person
    'Social Support': finland['Social support'],               # Level of social support
    'Life Expectancy': finland['Healthy life expectancy'],     # Average healthy lifespan
    'Freedom': finland['Freedom to make life choices'],        # Freedom to make life choices
    'Generosity': finland['Generosity'],                        # Generosity in society
    'Corruption': finland['Perceptions of corruption']         # Perceived corruption level
}
# Plot a pie chart to visualize the contributions of these factors
plt.pie(
    factors.values(),          # Values representing each factor's contribution
    labels=factors.keys(),     # Labels for each pie slice
    autopct='%1.1f%%',        # Display percentage contribution with 1 decimal place
    startangle=140            # Start the first slice at 140 degrees for better orientation
)
# Add a title to the plot
plt.title('Factor Contributions to Happiness Score (Finland)')
# Make sure pie chart is a circle by setting equal aspect ratio
plt.axis('equal')
# Show the pie chart
plt.show()
```

```python
# Select the row from the dataset corresponding to Finland
finland = dataset[dataset['Country or region'] == 'Finland'].iloc[0]
# Create a dictionary with the key factors contributing to Finland's happiness score
factors = {
    'GDP': finland['GDP per capita'],                          # Economic output per person
    'Social Support': finland['Social support'],               # Perceived social support level
    'Life Expectancy': finland['Healthy life expectancy'],     # Average healthy lifespan
    'Freedom': finland['Freedom to make life choices'],        # Freedom to make life decisions
    'Generosity': finland['Generosity'],                        # Willingness to donate to charity
    'Corruption': finland['Perceptions of corruption']         # Perceived level of corruption (lower is better)
}
# Create an explode list to offset each pie slice slightly for emphasis
explode = [0.05] * len(factors)  # 0.05 offset for each factor
# Plot an exploded pie chart with the factor contributions to happiness
plt.pie(
    factors.values(),         # Values to plot
    labels=factors.keys(),    # Corresponding labels for each slice
    autopct='%1.1f%%',       # Show percentage values with 1 decimal place
    startangle=140,          # Rotate the start angle for better visual appeal
    explode=explode,         # Offset each slice outward
    shadow=True              # Add a shadow effect for depth
)
# Set the title of the pie chart
plt.title('Exploded Pie Chart: Factor Contributions to Happiness Score (Finland)')
# Ensure the pie chart is a circle (equal aspect ratio)
plt.axis('equal')
# Display the plot
plt.show()
```

```python
# BOX PLOT
# Set the default plot style
plt.style.use('default')
# Extract the Happiness Score for Finland
finland = dataset[dataset['Country or region'] == 'Finland']['Score']
# Extract the Happiness Score for India
india = dataset[dataset['Country or region'] == 'India']['Score']
# Labels for the x-axis ticks
labels = ['Finland', 'India']
# Create a box plot to compare Happiness Scores of Finland and India
plt.boxplot([finland, india], tick_labels=labels)  # Use 'tick_labels' instead of deprecated 'labels'
# Add a title to the plot
plt.title('Generosity Comparison: Finland vs India')
# Label the y-axis
plt.ylabel('Generosity')
# Display the plot
plt.show()
```

```python
# Select top 5 countries by happiness score
top5 = dataset.sort_values(by='Score', ascending=False).head(5) # Sort in descending order and pick top 5
# Plotting a pie chart of their happiness scores
plt.pie(
    top5['Score'],                               # Values to plot
    labels=top5['Country or region'],            # Labels for each slice
    autopct='%1.1f%%',                           # Show percentage on slices
    startangle=140                               # Start angle for first slice
)
# Add a title and set aspect ratio to equal (to make it circular)
plt.title('Top 5 Countries Share of Total Happiness Score')
plt.axis('equal')  # Ensures the pie chart is a perfect circle
# Display the chart
plt.show()
```
